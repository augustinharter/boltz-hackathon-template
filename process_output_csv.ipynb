```python
import json
import os
from os import path
import pathlib
import shutil

import pandas as pd
```

```python
def unite_combined_results(csv_path_list):
    df_list = []
    for c in csv_path_list:
        local_df = pd.read_csv(c)
        id_name = path.basename(c).split(".")[0]
        print(id_name)
        local_df["run_id"] = [id_name] * len(local_df)
        df_list.append(local_df)
    df_combined = pd.concat(df_list, ignore_index=True)
    return df_combined
```

```python
csv_dir = "boltz_testruns"
csv_path_list = [
    path.join(csv_dir, f) for f in os.listdir(csv_dir) if ".csv" in f
]
```

```python
df_combined = unite_combined_results(csv_path_list)
```

**Note:** Relevant to extract confidence `.json` files

```python
def fetch_confidence_json(input, output_path, pattern="*.json"):
    in_path = pathlib.Path(input)
    in_json_paths = [
        p for p in in_path.rglob(pattern) 
    ]
    os.makedirs(output_path, exist_ok=True)
    for i in in_json_paths:
        shutil.copy(i, pathlib.Path(output_path) / i.name)
```

```python
fetch_confidence_json(
    "../raw_files_m_boltz_hackathon/ali_testrun/tmp", 
    "boltz_testruns/ali_confidence"
)
```
